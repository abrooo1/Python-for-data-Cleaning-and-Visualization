```python
import pandas as pd 
```

```python
df_r36_35 = pd.read_excel(r'Ethiopia DTM - SA R36 and R35 Consolidated Dataset_Without South Ethiopia.xlsx', sheet_name='Repr Sampling Demo')
```

```python
df_r36_35_site = pd.read_excel(r'Ethiopia DTM - SA R36 and R35 Consolidated Dataset_Without South Ethiopia.xlsx')
```

```python
df_r37 = pd.read_excel(r'Ethiopia DTM - SA R37 Dataset.xlsx')
```

```python
ki_rs_difference = pd.read_excel(r'Final Population estimate For SA round 37.xlsx')
```

```python
df_HH_average = df_r36_35.groupby('_uuid')['Number of member of HH'].mean()
```

```python
df_HH_average.shape
```

```python
df35_36_merged_with_site = df_r36_35_site[['1.1.c.1: Site ID', '_uuid','2.1.a.1: Total Number of IDP HHs', '2.1.b.7: Total Number of IDP Individuals','2.1.h.1: Number of Males by age: 0-4 years',
                                           '2.1.h.2: Number of Males by age: 5-14 years', '2.1.h.3: Number of Males by age: 15-17 years',
                                           '2.1.h.4: Number of Males by age: 18-59 years', '2.1.h.5: Number of Males by age: 60+ years',
                                           '2.1.i.1: Number of Female by age: 0-4 years', '2.1.i.2: Number of Female by age: 5-14 years', '2.1.i.3: Number of Female by age: 15-17 years',
                                           '2.1.i.4: Number of Female by age: 18-59 years', '2.1.i.5: Number of Female by age: 60+ years']].merge(df_HH_average,
                                       on= '_uuid', how='inner')
```

```python
df_r37.columns.to_list()
```

```python
df35_36_37_merged = df35_36_merged_with_site.merge(df_r37[['1.1.c.1: Site ID','2.1.a.1: Total Number of IDP HHs_R37', '2.1.b.7: Total Number of IDP Individuals_R37','M1515: Have there been new arrivals at the site during last four months?','uuu']],
                                       on= '1.1.c.1: Site ID', how='inner')
```

```python
df35_36_37_merged.columns
```

```python
df35_36_37_merged['Estimated_IND_R37_from_RS'] = (df35_36_37_merged['Number of member of HH'] * df35_36_37_merged['2.1.a.1: Total Number of IDP HHs_R37']).round(0)
```

```python
df35_36_37_merged
```

```python
df35_36_37_final= df35_36_37_merged.drop_duplicates(subset= '1.1.c.1: Site ID')
```

```python
df35_36_37_final.shape
```

```python
df35_36_37_final
```

```python
df35_36_37_final.shape
```

```python
df35_36_37_final= df35_36_37_final.copy()
df35_36_37_final['Difference_ki_DDC'] = abs(df35_36_37_final['Estimated_IND_R37_from_RS'] - df35_36_37_final['2.1.b.7: Total Number of IDP Individuals_R37'])
df35_36_37_final['10% of ColumR37'] = df35_36_37_final['Estimated_IND_R37_from_RS'] * 0.10
df35_36_37_final['Within ±10R37'] = df35_36_37_final['Difference_ki_DDC'] < df35_36_37_final['10% of ColumR37']
```
