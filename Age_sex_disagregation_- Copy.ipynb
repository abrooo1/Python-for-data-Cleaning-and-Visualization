```python
import pandas as pd
```

```python
df_r36_35 = pd.read_excel(r'Ethiopia DTM - SA R36 and R35 Consolidated Dataset_Without South Ethiopia.xlsx')
```

```python
df_r37 = pd.read_excel(r'Ethiopia DTM - SA R37 Dataset.xlsx')
```

```python
ki_rs_difference = pd.read_excel(r'Final Population estimate For SA round 37.xlsx')
```

```python
df_merged= df_r37[['1.1.c.1: Site ID', '2.1.a.1: Total Number of IDP HHs_R37', '2.1.b.7: Total Number of IDP Individuals_R37']].merge(df_r36_35[['1.1.c.1: Site ID','2.1.a.1: Total Number of IDP HHs','2.1.b.7: Total Number of IDP Individuals']], on = '1.1.c.1: Site ID', how='inner')
```

```python
df_merged.shape
```

```python
df_merged['Difference'] = abs(df_merged['2.1.a.1: Total Number of IDP HHs_x'] - df_merged['2.1.a.1: Total Number of IDP HHs_y'])
df_merged['10% of Column1'] = df_merged['2.1.a.1: Total Number of IDP HHs_x'] * 0.10
df_merged['Within ±10%'] = df_merged['Difference'] > df_merged['10% of Column1']
```

```python
df_merged['Within ±10%'].value_counts()
```

```python
df_333 = df_merged.merge(ki_rs_difference[['1.1.c.1: Site ID','Difference  between estimated Ind Vs KI individual ', 'Resampled_total_pop']], on='1.1.c.1: Site ID', how='inner')
```
