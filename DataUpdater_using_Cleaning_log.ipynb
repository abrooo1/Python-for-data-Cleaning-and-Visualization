```python
# Define the function to get the top three most reported columns
def top_three_most_reported_columns(row):
    top_values = row.value_counts().nlargest(3)
    top_columns = [row.index[row == value].tolist()[0] for value in top_values.index]
    return pd.Series(top_columns + [None] * (3 - len(top_columns)))  # Ensure three columns are returned

# Apply the function across rows and assign the output to three new columns in the DataFrame
barrier[['Top_Column_1', 'Top_Column_2', 'Top_Column_3']] = barrier.apply(top_three_most_reported_columns, axis=1)
```
