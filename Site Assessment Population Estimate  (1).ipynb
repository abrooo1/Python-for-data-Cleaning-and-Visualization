```python
def adjust_age_categories(df, age_cols, diff_col):
  
    def adjust_row(row):
        diff = row[diff_col]
        
        # Ensure the operation checks for non-zero float values
        non_zero_age_cols = [col for col in age_cols if row[col] != 0.0]

        if diff in [-1, -2, -3]:
            n = int(-diff)  # Cast diff to integer
            # Get the indices of the n smallest non-zero values
            min_indices = row[non_zero_age_cols].nsmallest(n).index
            row[min_indices] += 1.0  # Increment by 1.0 to ensure float operation
        elif diff in [1, 2, 3]:
            n = int(diff)  # Cast diff to integer
            # Get the indices of the n largest non-zero values
            max_indices = row[non_zero_age_cols].nlargest(n).index
            row[max_indices] -= 1.0  # Decrement by 1.0 to ensure float operation
        
        return row

    # Apply the adjustment row-wise
    df = df.apply(adjust_row, axis=1)
    return df
    
df = pd.read_excel(r'Representative Sampling Before normalization.xlsx',
    usecols=['Female 0-4_converted', 'Female 15-17_converted', 'Female 18-59_converted', 'Female 5-14_converted', 'Female Above 59_converted',
                                                         'Male 0-4_converted',  'Male 15-17_converted', 'Male 18-59_converted', 'Male 5-14_converted',
                                                         'Male Above 59_converted','diff', '_submission__uuid','diff'],
    index_col='_submission__uuid'
)

adjusted_df = adjust_age_categories(df, ['Female 0-4_converted', 'Female 15-17_converted', 'Female 18-59_converted', 'Female 5-14_converted', 'Female Above 59_converted',
                                                         'Male 0-4_converted',  'Male 15-17_converted', 'Male 18-59_converted', 'Male 5-14_converted',
                                                         'Male Above 59_converted'], 'diff')
```
